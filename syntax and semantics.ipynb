### Type inference
The type of python variables is infered at runtime, unlike other programming languages where you have to specify the type at compile time.

```python
dog = "Bobby"
## dog is a String

number = 10
## number is an integer

floater = 10.5
## floater is a float

print(type(dog))
print(type(number))
print(type(floater))
```

The type of variables is also dynamic, as it can be overriden

```python
one = 1
one = "one"

print(type(one))
```
